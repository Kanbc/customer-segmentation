## Create a segmented RFM table

```python
segmented_lfm = lfm[['acc_no','RB_L','RB_F','RB_M']].copy()
```

```python
rb0_quantiles['RB_L'][0.75]
```

```python
# ค่ามีความสัมพันธ์ไปทางเดียวกัน ยิ่งมากยิ่งดี จะได้
def LFMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
```

```python
segmented_lfm['L_quartile'] = segmented_lfm['RB_L'].apply(LFMScore, args=('RB_L',rb0_quantiles))
segmented_lfm['F_quartile'] = segmented_lfm['RB_F'].apply(LFMScore, args=('RB_F',rb0_quantiles))
segmented_lfm['M_quartile'] = segmented_lfm['RB_M'].apply(LFMScore, args=('RB_M',rb0_quantiles))
```

```python
segmented_lfm['LFMScore'] = segmented_lfm['L_quartile'].map(str) + segmented_lfm['F_quartile'].map(str) + segmented_lfm['M_quartile'].map(str)
segmented_lfm
```

```python
segmented_lfm.to_csv('output/LfmScore.csv', index=False)
```

```python
def lifetimeLabel(x):
    if x == 1:
        return "1 : ไม่เกิน 37 วัน"
    elif x == 2:
        return "2 : 38-90 วัน"
    elif x == 3:
        return "3 : 91-135 วัน"
    else:
        return "4 : มากกว่า 135 วัน"

def freqLabel(x):
    if x == 1:
        return "1 : 1 ครั้ง"
    elif x == 2:
        return "2 : 2 ครั้ง"
    elif x == 3:
        return "3 : 3-5 ครั้ง"
    else:
        return "4 : มากกว่า 5 ครั้ง"
    
def moneyLabel(x):
    if x == 1:
        return "1 : ไม่เกิน 1,000 บาท"
    elif x == 2:
        return "2 : 1,001-7,000 บาท"
    elif x == 3:
        return "3 : 7,001-15,261.8 บาท"
    else:
        return "4 : มากกว่า 15,261.8 บาท"
```

```python
tsne_data2 = pd.read_csv('output/tnse_data.csv')
tsne_data2['L_quartile'] = tsne_data2['L_quartile'].apply(lifetimeLabel)
tsne_data2['F_quartile'] = tsne_data2['F_quartile'].apply(freqLabel)
tsne_data2['M_quartile'] = tsne_data2['M_quartile'].apply(moneyLabel)
```

```python
tsne_data2.to_csv('output/tnse_data2.csv',index=False)
```
